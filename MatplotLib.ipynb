```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y) 
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
# Basic Plot
```

# Plotting more than one plot on one execution

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Create a figure
fig = plt.figure()

#Placed a list of axes
axes = fig.add_axes([0.1,0.1,0.8,0.8])

#plotted the data
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)

axes2.plot(y,x)
```

# WHat this shows is you can ITEARTE over subplots 

```python
fig, axes = plt.subplots(nrows = 1, ncols =2)

for current_ax in axes:
    current_ax.plot(x,y)
```
