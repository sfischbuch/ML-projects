# Categorical Plots

```python
import seaborn as sns
%matplotlib inline
tips.head()
```

```python
import numpy as np
```

```python
# The graph is with categroeis - sex vs. total bill
# The ESTIMATOR allows 
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

# Box plots

```python
# we see each total bill PER DAY 
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
# we see each total bill PER DAY- YOU CAN NOW SEE If the person is a smoker
# HUE IS SUPER AWESOME!

sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
# in this case ads JITTER - to see the 'stack points' 
sns.stripplot(x='day', y = 'total_bill', data=tips, jitter = True)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip',data=tips, col='sex', row = 'time')
```
